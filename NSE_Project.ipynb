```python
NSE_Dataset = pd.read_csv('/content/Equity Data (1 Jan 2020).csv.zip')
print(NSE_Dataset.columns)
NSE_Dataset=NSE_Dataset.drop([" OPEN_PRICE"," HIGH_PRICE"," LOW_PRICE"," LAST_PRICE"," AVG_PRICE"," TTL_TRD_QNTY"," TURNOVER_LACS"," NO_OF_TRADES"," DELIV_QTY"," DELIV_PER"], axis=1)
NSE_Dataset["change"]=(((NSE_Dataset[" CLOSE_PRICE"]-NSE_Dataset[" PREV_CLOSE"])/NSE_Dataset[" PREV_CLOSE"])*100)
```

```python
NSE_Dataset[' DATE1'] = pd.to_datetime(NSE_Dataset[' DATE1'], dayfirst=True)
NSE_Dataset['month'] = pd.DatetimeIndex(NSE_Dataset[' DATE1']).month
NSE_Dataset['year'] = pd.DatetimeIndex(NSE_Dataset[' DATE1']).year
NSE_Dataset=NSE_Dataset.sort_values(by=["SYMBOL","month","year"])
```

```python
### The query groups the stock based on particular month of the year and calculate monthly percentage change in the stock.
### hich is 
from pandasql import sqldf
q="select SYMBOL,month,year,avg(change) as change from df group by symbol,year,month order by SYMBOL"
newdf=sqldf(q, globals())
newdf
```

```python
newdf=newdf.sort_values(by=["year","month","change"])
newdf
```

# Monthly Top performing stocks

```python
q1="select symbol,month,year,max(change) from newdf group by year,month"
sqldf(q1, globals())
```

# Monthly worst performing stocks

```python
q2="select symbol,month,year,min(change) from newdf group by year,month"
sqldf(q2, globals())
```
